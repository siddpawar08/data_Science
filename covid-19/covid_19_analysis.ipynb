```python
df.sort_values(by = 'Confirmed',ascending = False)
```

Which country has the highest no of death cases all over ?


```python
max_cases = df.groupby("Country/Region")["Confirmed"].sum().sort_values(ascending = False).head(5)
```

We can see clearly that US has the maximum no of confirmed
 cases allover due to covid-19

```python
import matplotlib.pyplot as plt
plt.bar(max_cases.index, max_cases.values)
plt.xlabel("countries")
plt.ylabel("No of cases")
plt.show()
```

which country has the maximum no of death all over ?

```python
max_death = df.groupby("Country/Region")["Deaths"].sum().sort_values(ascending = False).head(5)
```

clearly , USA has the maximum no of deaths followed by UK , BRAZIL

```python
import matplotlib.pyplot as plt
plt.bar(max_death.index, max_death.values)
plt.xlabel("countries")
plt.ylabel("No of deaths")
plt.show()
```
