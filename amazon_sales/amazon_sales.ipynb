dropping rows with null values

```python
df.dropna(inplace = True)
```

changing datatype from object to float...  

```python
df['discounted_price'] = df['discounted_price'].str.replace("₹", "")
df['discounted_price'] = df['discounted_price'].str.replace(",", "")
df['discounted_price'] = df['discounted_price'].astype(float)
```

```python
df['actual_price']=df['actual_price'].str.replace("₹","")
df['actual_price']=df['actual_price'].str.replace(",","")
df['actual_price']=df['actual_price'].astype(float)
```

```python
df.head(5)
```

removing %  sign from discounted percentage

```python
df['discount_percentage']=df['discount_percentage'].str.replace("%","")
df['discount_percentage']=df['discount_percentage'].astype(float)


```

lets check which categories are sld the most

```python
import matplotlib.pyplot as plt
import seaborn as sns
category_top10 = df['category'].value_counts().head(10)
plt.figure(figsize = (10,6))
sns.barplot(x = category_top10.index , y = category_top10.values )
plt.xlabel("category ")
plt.ylabel("count")
plt.xticks(rotation = 45)
plt.show()
```

We can see clearly that computer accessories are sold the most , followed by smartphones (mobile accessories)

```python
df['rating_count']=df['rating_count'].str.replace(',','')
df['rating_count']=df['rating_count'].astype(int)
```

lets sort the data on basis of rating count

```python
import seaborn as sns
import matplotlib.pyplot as plt
rating_count = df.sort_values(by = 'rating_count' ,ascending = False)
plt.figure(figsize=(10,6))
sns.barplot(x = rating_count['product_name'].head(10) , y = rating_count['rating_count'].head(10) ,color ='maroon' , )
plt.xlabel('product name')
plt.ylabel('rating_count')
plt.show()
```
